```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Load Data

```python
df = pd.read_csv("osteoporosis.csv")
df
```

```python
import matplotlib.pyplot as plt

# Mengambil data jumlah individu yang terkena osteoporosis per rentang usia
jumlah_terkena_osteoporosis_per_umur = df[df["Osteoporosis"] == 1].groupby("Age")["Osteoporosis"].count()

# Membuat plot grafik batang
plt.figure(figsize=(15, 6))
jumlah_terkena_osteoporosis_per_umur.plot(kind='bar', color='skyblue')
plt.title('Number of Individuals Affected by Osteoporosis per Age Range')
plt.xlabel('Age')
plt.ylabel('Number Affected by Osteoporosis')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# Memfilter data berdasarkan rentang usia 18-90
df_filtered = df[(df['Age'] >= 18) & (df['Age'] <= 90)]

# Menghitung jumlah kasus osteoporosis untuk setiap sesi usia
jumlah_kasus_sesi_1 = df_filtered[(df_filtered['Age'] >= 18) & (df_filtered['Age'] <= 30) & (df_filtered['Osteoporosis'] == 1)].shape[0]
jumlah_kasus_sesi_2 = df_filtered[(df_filtered['Age'] >= 31) & (df_filtered['Age'] <= 50) & (df_filtered['Osteoporosis'] == 1)].shape[0]
jumlah_kasus_sesi_3 = df_filtered[(df_filtered['Age'] >= 51) & (df_filtered['Age'] <= 70) & (df_filtered['Osteoporosis'] == 1)].shape[0]
jumlah_kasus_sesi_4 = df_filtered[(df_filtered['Age'] >= 71) & (df_filtered['Age'] <= 90) & (df_filtered['Osteoporosis'] == 1)].shape[0]

# Menampilkan jumlah kasus osteoporosis untuk setiap sesi usia
print("Jumlah kasus osteoporosis pada rentang usia 18-30: ", jumlah_kasus_sesi_1)
print("Jumlah kasus osteoporosis pada rentang usia 31-50: ", jumlah_kasus_sesi_2)
print("Jumlah kasus osteoporosis pada rentang usia 51-70: ", jumlah_kasus_sesi_3)
print("Jumlah kasus osteoporosis pada rentang usia 71-90: ", jumlah_kasus_sesi_4)
```

# Spliting Data

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report,confusion_matrix
from sklearn.naive_bayes import GaussianNB
```

```python
X = df[["Age","Gender","Hormonal Changes",'Family History','Race/Ethnicity','Body Weight','Calcium Intake','Vitamin D Intake','Physical Activity','Smoking','Alcohol Consumption','Medical Conditions','Medications','Prior Fractures']]
y = df["Osteoporosis"]
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
model = GaussianNB()
model.fit(X_train,y_train)
pred = model.predict(X_test)
report = classification_report(y_test,pred)
print("Naive Bayes")
print(report)
```

```python
from sklearn.ensemble import GradientBoostingClassifier
modelGBC = GradientBoostingClassifier(random_state=42)
modelGBC.fit(X_train,y_train)
predGBC = modelGBC.predict(X_test)
reportGBC = classification_report(y_test,predGBC)
print("gradient Boosting Clasifier")
print(reportGBC)
conf_mx_GBC = confusion_matrix(y_test,predGBC)
plt.figure(figsize=(10,10))
sns.heatmap(conf_mx_GBC,annot=True)
```

```python
import joblib
joblib.dump(modelGBC,'ModelGBC.joblib')
```

```python
from sklearn.svm import SVC
modelSVC = SVC(random_state=42)
modelSVC.fit(X_train,y_train)
predSVC = modelSVC.predict(X_test)
reportSVC = classification_report(y_test,predSVC)
print("Super Vector Clasifier")
print(reportSVC)
```

```python
from sklearn.neighbors import KNeighborsClassifier
modelKNN = KNeighborsClassifier(n_neighbors=5)
modelKNN.fit(X_train,y_train)
predKNN = modelKNN.predict(X_test)
reportKNN = classification_report(y_test,predKNN)
print("K-NN")
print(reportKNN)
```

```python
from sklearn.linear_model import SGDClassifier
modelSGD = SGDClassifier()
modelSGD.fit(X_train,y_train)
predSGD = modelSGD.predict(X_test)
reportSGD = classification_report(y_test,predSGD)
print(reportSGD)
```

```python
from xgboost import XGBClassifier
modelXGB = XGBClassifier()
modelXGB.fit(X_train,y_train)
predXGB = modelXGB.predict(X_test)
reportXGB = classification_report(y_test,predXGB)
print(reportXGB)
```

```python
from catboost import CatBoostClassifier
modelCB = CatBoostClassifier(random_state=42)
modelCB.fit(X_train,y_train)
predCb = modelCB.predict(X_test)
reportCB = classification_report(y_test,predCb)
print(reportCB)
conf_mx_CB = confusion_matrix(y_test,predCb)
plt.figure(figsize=(10,10))
sns.heatmap(conf_mx_CB,annot=True)
```
